# Data Mining: Used Cars on Craigslist

```python
import os
import zipfile
import pandas as pd
```

## Download Data

```python
%%bash
kaggle datasets download -d austinreese/craigslist-carstrucks-data -p data
```

```python
def fetch_cars_data():
    zip_path = os.path.join('data', 'craigslist-carstrucks-data.zip')    
    
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall('data')
    
    print(zip_ref.namelist())

fetch_cars_data()
```

## Load Data

```python
def load_data():    
    csv_path = os.path.join('data', 'vehicles.csv')
    return pd.read_csv(csv_path)

vehicles = load_data()
```
